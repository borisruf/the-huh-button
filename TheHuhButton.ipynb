```python
# add punctuation to video transcripts

import os
import openai

openai.api_type = "XXX"
openai.api_base = "XXX"
openai.api_version = "XXX"
openai.api_key = "XXX"

folder_path = "data/raw"
transcripts = {}

def truncate_text(text, max_token):
    words = text.split()
    chunks = [' '.join(words[i:i+max_token]) for i in range(0, len(words), max_token)]
    return chunks

for file in os.listdir(folder_path):
    if file.endswith(".txt"):
        file_path = os.path.join(folder_path, file)
        with open(file_path, 'r') as f:
            transcripts[file] = f.read()


for filename, transcript in transcripts.items():

    punctuated_transcript = ""

    # break up the text into smaller parts
    chunks = truncate_text(transcript, 300)

    # limit the number of requests to save cost/energy - a shorter text is sufficient for demonstration purposes
    chunks = chunks[:10]

    for chunk in chunks:
        response = openai.ChatCompletion.create(
        
            engine = "XXX",
            messages = [
                {"role": "system", "content": "Add punctuation to this text:\n\n" + chunk}
            ]
        )
        
        punctuated_transcript += " " + response.choices[0]['message']['content']

    # write punctuated text to new folder
    with open("data/punctuated/" + filename, 'w') as file:
        file.write(punctuated_transcript)
```

```python
# load transcripts

import os

folder_path = "data/punctuated"
text_files = []

for file in os.listdir(folder_path):
    if file.endswith(".txt"):
        file_path = os.path.join(folder_path, file)
        with open(file_path, 'r') as f:
            text_files.append(f.read())
```

```python
# Randomly select positions in the transcript histories (the moment, the user sends the signal/hits the button),
# allowing for a certain margin at the beginning

import random

histories = []

for text in text_files:

    if len(text) > 100:
        random_index = random.randint(500, len(text) - 1)
        histories.append(text[:random_index])
    else:
        histories.append(text)
```
